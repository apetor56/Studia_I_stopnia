## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:
plt.scatter(file['sqft_living'], file['price'])
plt.xlabel('square footage')
plt.ylabel('price')

#Hint plt.scatter
```

There is weak correlation between square footage and price.

```python
#b)
#Your code:
plt.scatter(file['grade'], file['price'])
plt.xlabel('grade')
plt.ylabel('price')
```

```python
#c)
#Your code:
plt.scatter(file['zipcode'], file['price'])
plt.xlabel('zipcode')
plt.ylabel('price')
```

```python
#Your code:
plt.scatter(file['long'], file['lat'], s=10, c=file['grade'])
plt.xlabel('longitude')
plt.ylabel('lattitude')
```

Most expensive apartments are located around city center.

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
selected = file.loc[(file['bedrooms'] == 3) & (file['bathrooms'] == 2) & (file['floors'] == 2)]
selected.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.scatter(file.grade, file.price)
plt.scatter(selected.grade, selected.price)
plt.xlabel("grade")
plt.ylabel("price")

```

Apartments we are searching for have grade point range from 6 to 10.

```python
#c)
#Your code:
selected['ratio'] = selected['price'] / selected['sqft_living']
selected.sort_values('ratio', ascending=True)

#Hint: .loc, .sort_values(), .index
```

As we can see apartment with best given ratio (95.940959, which is the smallest) has id equal 9704.
